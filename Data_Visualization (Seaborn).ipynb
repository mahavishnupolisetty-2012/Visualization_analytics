## Plotting Categorical Data

#### Jitter, Hue, Boxplot, Voilin Plot, Pointplot


#### In the above section, we saw how we can use different visual representations to show the relationship between multiple variables. We drew the plots between two numeric variables. In this section, we’ll see the relationship between two variables of which one would be categorical (divided into different groups).



#### We’ll be using catplot() function of seaborn library to draw the plots of categorical data. Let’s dive in

## Jitter Plot


```python
sns.relplot(x="CarName", y="carbody", data = df) 
```

```python
sns.catplot(x="CarName", y="carbody", jitter = False,  data = df) 
```

## Hue Plot

```python
sns.catplot(x="CarName", y="carbody", hue = "doornumber", data = df)  #Next, if we want to introduce another variable or another dimension in our plot, we can use the hue parameter just like we used in the above section.
```

```python
sns.catplot(x="CarName", y="carbody", kind = "swarm", data = df)  #kind = “swarm”, swarm uses an algorithm that prevents the points from overlapping and adjusts the points along the categorical axis. 
```

```python
sns.catplot(x="CarName", y="carbody", hue = "doornumber", kind = "swarm", data = df)
```

## Boxplot using seaborn

```python
sns.catplot(y="wheelbase", x="symboling", kind = "box", data = df) # Each value in the boxplot corresponds to actual observation in the data.
```

```python
sns.catplot(y="wheelbase", x="symboling", hue = "doornumber", kind = "box", data = df) # When we use hue semantic with boxplot, it is leveled along the categorical axis so they don’t overlap.
```

## Violin Plot using seaborn

#### The violin plots combine the boxplot and kernel density estimation procedure to provide richer description of the distribution of values. 

```python
sns.catplot(y="wheelbase", x="symboling", hue = "doornumber", kind = "violin", data = df) ## The quartile values are displayed inside the violin.
```

##### We can also split the violin when the hue semantic parameter has only two levels, which could also be helpful in saving space on the plot. 

```python
sns.catplot(y="wheelbase", x="symboling", hue = "doornumber", kind = "violin", split = True, data = df)
```

## Boxplot using seaborn

##### Boxplot operates on the full dataset and obtains the mean value by default.

```python
sns.catplot(y="wheelbase", x="carbody", hue = "doornumber", kind = "bar", data = df)
```

## pointplot

##### this plot points out the estimate value and confidence interval. Pointplot connects data from the same hue category. This helps in identifying how the relationship is changing in a particular hue category.

```python
sns.catplot(y="wheelbase", x="carbody", hue = "doornumber", kind = "point", data = df)
```

### Swarm plot using seaborn

#### using this we can visualize higher dimension relationships as well. 

```python
sns.catplot(x="carbody", y="wheelbase",  hue = "doornumber",
           col = "fueltype", aspect=0.9,
           kind="swarm", data=df);
```

### Visualizing the Distribution of a Dataset

##### Whenever we are dealing with a dataset, we want to know how the data or the variables are being distributed. Distribution of data could tell us a lot about the nature of the data, so let’s dive into it.

#### Histogram

```python
sns.distplot(df.wheelbase) #distplot() function draws histogram and fits a Kernel Density Estimate.
```

```python
sns.distplot(df.wheelbase, kde=False, rug = True)
```

## Plotting Bivariate Distributions

##### Hexplot, KDE plot, Boxen plot, Ridge plot (Joyplot)

#### Bivariate means joint, so to visualize it, we use jointplot() function of seaborn library. By default, jointplot draws a scatter plot.

```python
sns.jointplot(x="wheelbase", y="symboling", data=df);
```

## Hexplot using Seaborn

#### it is a bivariate analog of histogram as it shows the number of observations that falls within hexagonal bins. This is a plot which works with large dataset very easily. To draw a hexplot, we’ll set kind attribute to hex. 

```python
sns.jointplot(x="wheelbase", y="symboling", kind = "hex", data=df);
```

## KDE Plot using Seaborn

### It’s another method to visualize the bivariate distribution. above observations could also be achieved by using jointplot() function and setting the attribute kind to KDE.

```python
sns.jointplot(x="wheelbase", y="symboling", data=df, kind = "kde");
```

### Heatmaps using Seaborn

##### Heatmaps are graphical representations in which each variable is represented as a color.

```python
corrmat = df.corr()
f,ax=plt.subplots(figsize=(9,6))
sns.heatmap(corrmat, vmax=8, square=True)
```

## Boxen Plot using Seaborn

##### use to show the bivariate distribution is boxen plot. Boxen plots were originally named letter value plot as it shows large number of values of a variable, also known as quantiles. These quantiles are also defined as letter values. By plotting a large number of quantiles, provides more insights about the shape of the distribution. These are similar to box plots.

```python
sns.jointplot(x="wheelbase", y="car_ID", data=df, kind = "boxen", height=4, aspect=2.7, hue="carbody");
```
